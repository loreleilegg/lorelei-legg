# Final Project: A comparison of Cases and Deaths due to COVID-19 in 3 highly populated countries: Japan, US and Italy

```python
%matplotlib inline
```

```python
import pandas
```

```python
df=pandas.read_excel('s3://legg-ia241/COVID-19-geographic-disbtribution-worldwide_April_10_2020.xlsx')
df
```

Imported data from The  European Centre for Disease Prevention and Control.

# Japan

```python
Japan_data = df.loc[df['countriesAndTerritories'] == 'Japan']
Japan_data
```

```python
Japan_data.groupby('dateRep').sum()['cases'].plot()
```

The amount of cases is increasing every month, with a few instances of decrease of major decreases followed by large spikes.

```python
Total_cases_Japan = df.loc[df['countriesAndTerritories'] == 'Japan'].sum()['cases']
Total_cases_Japan
```

```python
Japan_data.groupby('dateRep').sum()['deaths'].plot()
```

This data compares the amount of days to the amount of deaths, the graph shows an overall increase in deaths astime increases, with instances of major decreases and major spikes.

```python
TotalDeaths_Japan = df.loc[df['countriesAndTerritories'] == 'Japan'].sum()['deaths']
TotalDeaths_Japan
```
